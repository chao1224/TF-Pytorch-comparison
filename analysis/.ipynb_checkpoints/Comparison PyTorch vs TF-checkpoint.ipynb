```python
%matplotlib inline
import matplotlib.pyplot as plt
%pylab inline
pylab.rcParams['figure.figsize'] = (10, 8)
```

```python
TF_TRAIN_OUT_FILE = "train_out"
TF_EVAL_OUT_FILE = "eval_out" 
TORCH_OUT_FILE = 'out.txt'
```

```python
# fetch data
with open(TF_TRAIN_OUT_FILE, 'rb') as train_out:
    train_accuracy = []
    time_cost = []
    for line in train_out.readlines():
        tmp_line_content = line.strip('\n').split(',')
        epoch_idx = int(tmp_line_content[0].split('[')[0].split(":")[-1])
        if epoch_idx <= 63:
            train_accuracy.append(float(tmp_line_content[-1].split(':')[-1]))
            time_cost.append(float(tmp_line_content[-2].split(":")[-1]))
with open(TF_EVAL_OUT_FILE, 'rb') as eval_out:
    eval_accuracy = []
    step_index = []
    for line in eval_out.readlines():
        tmp_line_content=line.strip('\n').split(',')
        if len(eval_accuracy) <= 63:
            step_index.append(int(tmp_line_content[0].split(':')[-1]))
            eval_accuracy.append(1.0-float(tmp_line_content[2].split(':')[-1]))
with open(TORCH_OUT_FILE, 'rb') as torch_out:
    train_acc = []
    valid_acc = []
    valid_step = []
    time_cost_torch = []
    for line in torch_out.readlines():
        tmp_line_content = line.strip('\n').split(',')
        if tmp_line_content[0][0:5] == "Train":
            # training output
            train_acc.append(float(tmp_line_content[-1].split(':')[-1]))
            time_cost_torch.append(float(tmp_line_content[-2].split(':')[-1]))
        elif tmp_line_content[0][0:5] == "Epoch":
            # validation output
            valid_acc.append(1.0-float(tmp_line_content[2].split(':')[-1]))
            valid_step.append(int(tmp_line_content[1].split(':')[-1]))

print("Total Time cost for 63 epochs of TF is {}s".format(sum(time_cost)))
print
print("Total Time cost for 63 epochs of PyTorch is {}s".format(sum(time_cost_torch)))
```

```python
fig = plt.figure(1)
plt.plot([i for i in range(len(train_accuracy))], train_accuracy, 'r', label="TF train prec")
plt.plot(step_index, eval_accuracy, ':r', label='TF validation prec', linewidth="3.0")
plt.plot([i for i in range(len(train_acc))], train_acc, 'b', label="Torch train prec")
plt.plot(valid_step, valid_acc, ':b', label='Torch validation prec', linewidth="3.0")
plt.title("TF vs Pytorch Cifar10 on ResNet-32")
plt.legend()
plt.show()
```
