```python
%matplotlib inline
import matplotlib.pyplot as plt
%pylab inline
pylab.rcParams['figure.figsize'] = (10, 8)
```

```python
arange = [1, 2, 4, 8]
list_0 = [1, 1.812, 3, 1.25]
list_1 = [1, 2, 3.3333, 2.3809]
list_2 = [1, 1.94, 3.46428, 4.85]
list_3 = [1, 1.7776, 3.222, 5.172413793]
```

This experiment is runining on m4.2xlarge instance of AWS EC2. The dataset is MNIST and netowrk used here is LeNet, I tried different batch size from 512 to 4096 and observe the speed up wrt time cost of data pass.

```python
fig = plt.figure(1)
plt.plot(arange, list_0, '-vc', label="batch-size: 512")
plt.plot(arange, list_1, '-^r', label="batch-size: 1024")
plt.plot(arange, list_2, '-sg', label="batch-size: 2048")
plt.plot(arange, list_3, '-ob', label="batch-size: 4096")
plt.xlabel("number of workers")
plt.ylabel("speed up")
plt.legend()
plt.title("MIST on Distributed LeNet")
plt.show()
```
